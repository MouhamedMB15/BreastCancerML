```python
# Importing Dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt 

```

```python

# Data Collection & Processing
# Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)

```

```python
#Loading the Data to a data frame 
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
#Test pringing the first 5 rows of the dataframe 
data_frame.head()
```

```python
# Adding the target columns to the data frame 
data_frame['label'] = breast_cancer_dataset.target
#Printing results of the last 5 rows in data frame
data_frame.tail()
```

```python
#Anyalisis of data / number of rows and colums in the data set
data_frame.shape
```

```python
#569 distinc rows and 31 distinct colums 
#Getting info based on the data
data_frame.info()
```

```python
#Determing if any missing datasets
data_frame.isnull().sum
```

```python
#Statisitical Measures about the data set
data_frame.describe()
```

```python
# Checking the distrubition of target variables 
data_frame['label'].value_counts()
```

```python
#1 repo the benigin cases 0 -- repo the maligant cases 
#Group bases on data set
data_frame.groupby('label').mean()
```

```python
#Seperating the Features and target
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
print(X)
```

```python
print(Y)
```

```python
#Disitncign data with training data and testing data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Standarize the data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
#Model training 
#Logisistic Regression
model = LogisticRegression()
# Training the logistic regression model using Training Data 
model.fit(X_train, Y_train)
```

```python
#Model Evaulation 
#Accucracy Score 
# Accuracy on training data train data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

print('Accuracy on training data = ' , training_data_accuracy)

```

```python
# Accuracy on training data test
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)

print('Accuracy on testing data = ' , test_data_accuracy)
```

```python
#importing tensorflow
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras

```

```python
#Setting layers for nueral network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #input
    keras.layers.Dense(20, activation='relu'), #inner
    keras.layers.Dense(2, activation='sigmoid') #output
])
```

```python
#complining the neueral network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
#training the nueral network 

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
#Visualizing the accuracy data

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
#Vsisualizing loss data

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
#Accuracy of the model on test data check

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python

#Creating Backend model

import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform



# Standardize the training data
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

# Logistic Regression model
model = LogisticRegression()

# Params for random search
param_dist = {
    'C': uniform(0.1, 10),
    'solver': ['liblinear', 'saga']
}

# Random search
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, scoring='accuracy', cv=5, n_jobs=-1, random_state=42)
random_search.fit(X_train_std, Y_train)

# Return best params and best score
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

# Save the best model
best_model = random_search.best_estimator_
joblib.dump(best_model, 'model.joblib')

# Save the scaler
joblib.dump(scaler, 'scaler.joblib')





```
